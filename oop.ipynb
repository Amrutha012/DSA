```python
class Student:
  def __init__(self,name,student_id):
    self.name = name
    self.student_id = student_id 
    self.subjects={} #subjects:marks dictionary

  def add_subject(self,subject,marks):
    self.subjects[subject]=marks 

  def total_marks(self):
    return sum(self.subjects.values())
  
  def print_report_card(self):
    #print("             ",self.name,"               ")
    print("============== REPORT CARD ==============")
    print(f"Name       : {self.name}")
    print(f"Student ID : {self.student_id}")
    print("-----------------------------------------")
    for subject,marks in self.subjects.items():
      print(f"{subject:<10}: {marks}")
    print("-----------------------------------------")
    print(f"Total Marks: {self.total_marks()}")
    print("=========================================")

#Example Usage 
s1 = Student("Amrutha",101)
s1.add_subject("Python",95)
s1.add_subject("Java",90)
s1.add_subject("Maths",93)
s1.add_subject("DBMS", 89)

s1.print_report_card()
```

```python
#Abstract class
from abc import ABC, abstractmethod 

class ATM(ABC):
  @abstractmethod
  def withdraw(self,amount):
    pass

  @abstractmethod
  def deposit(self,amount):
    pass

  @abstractmethod
  def check_balance(self):
    pass

#concrete class
class SBI_ATM(ATM):
  def __init__(self,balance):
    self.balance = balance
  
  def withdraw(self,amount):
    if amount<=self.balance:
      self.balance-=amount
      print(f"Withdrawn: {amount}, Remaining amount: {self.balance}")
    else:
      print("Insufficient Balance!")

  def deposit(self,amount):
    self.balance+=amount
    print(f"Deposited: {amount}, Current balance: {self.balance}")

  def check_balance(self):
    print(f"Your Balance: {self.balance}")

#Using Abstraction
user1 = SBI_ATM(5000)

user1.check_balance()
user1.withdraw(2000)
user1.deposit(1000)
user1.check_balance()
```

```python
#Inheritance
class Parent:
  def work(self):
    print("Parent is working in the office.")

class Daughter(Parent): #Inheriting Parent
  def study(self):
    print("Daughter is studying in college.")


p = Parent()
p.work()

d = Daughter()
d.work() #Inherited from Parent
d.study() #Daughter's own method
```

```python
class Student:
  def __init__(self,name,marks):
    self.__name = name #private variable
    self.__marks = marks #private variable

  #getter for name 
  def get_name(self):
    return self.__name
  
  #getter for marks
  def get_marks(self):
    return self.__marks
  
  #setter for marks
  def set_marks(self,marks):
    if 0<= marks <=100:
      self.__marks = marks 
    else:
      print("Invalid marks! Must be from 0 to 100.")

#Usage 
s1 = Student("Amrutha",95)

print("Name:", s1.get_name())
print("Marks:", s1.get_marks())

s1.set_marks(100) #valid update
print("Updated Marks:", s1.get_marks())
```

```python
#polymorphism
class Calculator:
  def add(self,a,b=0,c=0):
    return a+b+c
  
calc = Calculator()
print(calc.add(10)) # 1 argument
print(calc.add(10,20)) # 2 arguments
print(calc.add(10,20,30)) # 3 arguments
```

```python
import time
class Hotel:
  def __init__(self,name,amount):
    self.__name = name
    self.__amount = amount 

  def get_name(self):
    return self.__name
  def get_amount(self):
    return self.__amount
  def payment_mode(self,mode):
    if mode ==1:
      return "You selected Card Payment"
    elif mode==2:
      return "You selected UPI Payment"
    elif mode==3:
      return "You selected Wallet Payment"
    else:
      return "Invalid Payment Mode"

h1 = Hotel("Amrutha",20000)
print(f"Hello {h1.get_name()}")
print(f"Your Wallet Balance is: {h1.get_amount()} ")

print("Select a payment mode:")
print("1.Card")
print("2.UPI")
print("3.Wallet")

choice = int(input("Enter your choice (1/2/3):"))
print(h1.payment_mode(choice))

if choice ==1:
  print("Enter CVV number:")
  cvv = int(input())

  print("Processing card payment...")
  time.sleep(3)

  if len(str(cvv))==3:
    print("Card payment processed")
  else:
    print("Invalid CVV Details")
elif choice==2:
  print("Please enter UPI ID:")
  upi = input()
  print("processing UPI Payment...")
  time.sleep(3)

  if '@' in upi:
    print("UPI Payment Successful")
  else:
    print("Invalid UPI ID")
elif choice==3:
  money = int(input("Enter amount needed:"))

  if money<=h1.get_amount():

    print("Processing Wallet Payment...")
    time.sleep(3)
    balance=h1.get_amount()-money
    print("Wallet Payment Successful")
    print(f"Remaining Wallet Balance: {balance}")
  else:
    print("Processing Wallet Payment...")
    time.sleep(3)
    print("Insufficient Wallet Balance")
```
