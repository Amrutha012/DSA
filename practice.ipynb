```python
#Binary search 
#First Occurence
def binary_search(arr,target):
    n = len(arr)
    start = 0
    end = n-1 
    ans = -1 
    
    
    while start<=end:
        mid = (start+end)//2 
        if arr[mid]==target:
            ans = mid
            end = mid-1
        elif arr[mid]>target:
            end = mid-1
        else:
            start = mid + 1 
            
    return ans 
    
sorted_list = [1,2,2,4,5,6]
target = 2

result = binary_search(sorted_list,target)
print(result)
```

```python
#Last Occurerence
def binary_search(arr,target):
    n = len(arr)
    start = 0
    end = n-1 
    ans = -1 
    
    
    while start<=end:
        mid = (start+end)//2 
        if arr[mid]==target:
            ans = mid
            start = mid+1
        elif arr[mid]>target:
            end = mid-1
        else:
            start = mid + 1 
            
    return ans 
    
sorted_list = [1,2,2,4,5,6]
target = 2

result = binary_search(sorted_list,target)
print(result)
```
